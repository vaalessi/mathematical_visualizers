```python
import matplotlib
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt



fig = plt.figure(figsize=(10, 10), dpi=80)
ax = plt.axes(projection='3d')
t = np.linspace(0, 1, 100)

x = t
y = 1/2-2*t
z = t
```

```python
ax.plot3D(x, y, z, 'red')
ax.set_title('3D line plot')
```
